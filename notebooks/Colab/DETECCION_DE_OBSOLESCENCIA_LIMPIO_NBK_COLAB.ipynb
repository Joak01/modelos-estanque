```python
tsp = TSP(0.001, 0.1, 0.0000750)

n = 0

#ventana fija
start = int((36000-3600*1)/10)
stop = int((360000*+3600*3)/10)

y = df['hc'][start:stop]
y_hat = df['y_hat_1'][start:stop]

y_fm1 = df['f_pid'][start:stop]
y_fm1 = (y_fm1-y_fm1.min())/(y_fm1.max()-y_fm1.min())

y_fm2 = df['freq_prev'][start:stop]
y_fm2 = (y_fm2-y_fm2.min())/(y_fm2.max()-y_fm2.min())


y_m1 = df['hc_prev'][start:stop]
y_m1 = (y_m1-y_m1.min())/(y_m1.max()-y_m1.min())

y_m2 = df['hc_prev_prev'][start:stop]
y_m2 = (y_m2-y_m2.min())/(y_m2.max()-y_m2.min())


e = y-y_hat
e = (e-e.min())/(e.max()-e.min())

e_tsp = np.copy(np.expand_dims(e, axis=1), order='F')

yfm1_tsp =  np.copy(np.expand_dims(y_fm1, axis=1), order='F')
yfm2_tsp =  np.copy(np.expand_dims(y_fm2, axis=1), order='F')

ym1_tsp = np.copy(np.expand_dims(y_m1, axis=1), order='F')
ym2_tsp = np.copy(np.expand_dims(y_m2, axis=1), order='F')
```

```python
tsp.grow(yfm1_tsp, e_tsp)
tsp.regularize()
emi_yfm1 = tsp.emi()

tsp.grow(yfm2_tsp, e_tsp)
tsp.regularize()
emi_yfm2 = tsp.emi()

tsp.grow(ym1_tsp, e_tsp)
tsp.regularize()
emi_ym1 = tsp.emi()

tsp.grow(ym2_tsp, e_tsp)
tsp.regularize()
emi_ym2 = tsp.emi()
```

```python
k=0
plt.figure(figsize=(20,16))
plt.subplot(8,4,k+1)
plt.plot(y_m1, e, '.', alpha=0.2, color='tab:blue')
plt.xlabel(r'$y(t-1)$')
plt.ylabel(r'$e$')
plt.plot([], [], ' ', label=r'$\hat{{I}}(y(t-1),e)={:.3f}$'.format(emi_ym1))
plt.legend(loc='upper left', handlelength=0)
plt.gca().set_aspect('equal', 'datalim')

plt.subplot(8,4,k+5)
plt.plot(y_m2, e, '.', alpha=0.2, color='tab:orange')
plt.xlabel(r'$y(t-2)$')
plt.ylabel(r'$e$')
plt.plot([], [], ' ', label=r'$\hat{{I}}(y(t-2),e)={:.3f}$'.format(emi_ym2))
plt.legend(loc='upper left', handlelength=0)
plt.gca().set_aspect('equal', 'datalim')   


plt.subplot(8,4,k+17)
plt.plot(y_fm1, e, '.', alpha=0.2, color='tab:orange')
plt.xlabel(r'$f$')
plt.ylabel(r'$e$')
plt.plot([], [], ' ', label=r'$\hat{{I}}(f(t),e)={:.3f}$'.format(emi_yfm1))
plt.legend(loc='upper left', handlelength=0)
plt.gca().set_aspect('equal', 'datalim')  

plt.subplot(8,4,k+21)
plt.plot(y_fm2, e, '.', alpha=0.2, color='tab:orange')
plt.xlabel(r'$f$')
plt.ylabel(r'$e$')
plt.plot([], [], ' ', label=r'$\hat{{I}}(f(t-1),e)={:.3f}$'.format(emi_yfm2))
plt.legend(loc='upper left', handlelength=0)
plt.gca().set_aspect('equal', 'datalim') 

```

```python
tsp = TSP(0.01, 0.1, 0.000000750)

n = 0
# DISCRETE_INPUT = True


window_size = int(3600/10*2) # 2 hours



start = 0
stop = -1

y = df['hc'][start:stop]
y_hat = df['y_hat_1'][start:stop]

y_fm1 = df['f_pid'][start:stop]
# y_fm1 = (y_fm1-y_fm1.min())/(y_fm1.max()-y_fm1.min())

y_fm2 = df['freq_prev'][start:stop]
# y_fm2 = (y_fm2-y_fm2.min())/(y_fm2.max()-y_fm2.min())

y_fm3 = y_fm1-y_fm2  #df['freq_prev_3'][start:stop]
# y_fm3 = (y_fm3-y_fm3.min())/(y_fm3.max()-y_fm3.min())

# y_fm4 = df['freq_prev_4'][start:stop]
# y_fm4 = (y_fm4-y_fm4.min())/(y_fm4.max()-y_fm4.min())


y_m1 = df['hc_prev'][start:stop]
# y_m1 = (y_m1-y_m1.min())/(y_m1.max()-y_m1.min())

y_m2 = df['hc_prev_prev'][start:stop]
# y_m2 = (y_m2-y_m2.min())/(y_m2.max()-y_m2.min())

# y_m3 = df['hc_prev_3'][start:stop]
# y_m3 = (y_m3-y_m3.min())/(y_m3.max()-y_m3.min())

# y_m4 = df['hc_prev_4'][start:stop]
# y_m4 = (y_m4-y_m4.min())/(y_m4.max()-y_m4.min())


e = y-y_hat


emis_fm1 = [None] * window_size
emis_fm2 = [None] * window_size
emis_fm3 = [None] * window_size
# emis_fm4 = [None] * window_size

emis_ym1 = [None] * window_size
emis_ym2 = [None] * window_size
# emis_ym3 = [None] * window_size
# emis_ym4 = [None] * window_size

for i in range(len(y)-window_size+1):
  # print(i)

  e_tsp = np.copy(np.expand_dims(e[i:i+window_size], axis=1), order='F')

  yfm1_tsp =  np.copy(np.expand_dims(y_fm1[i:i+window_size], axis=1), order='F')
  yfm2_tsp =  np.copy(np.expand_dims(y_fm2[i:i+window_size], axis=1), order='F')
  yfm3_tsp =  np.copy(np.expand_dims(y_fm3[i:i+window_size], axis=1), order='F')
  # yfm4_tsp =  np.copy(np.expand_dims(y_fm4[i:i+window_size], axis=1), order='F')


  ym1_tsp = np.copy(np.expand_dims(y_m1[i:i+window_size], axis=1), order='F')
  ym2_tsp = np.copy(np.expand_dims(y_m2[i:i+window_size], axis=1), order='F')
  # ym3_tsp = np.copy(np.expand_dims(y_m3[i:i+window_size], axis=1), order='F')
  # ym4_tsp = np.copy(np.expand_dims(y_m4[i:i+window_size], axis=1), order='F')

  #################################################

  tsp.grow(yfm1_tsp, e_tsp)
  tsp.regularize()
  emi_yfm1 = tsp.emi()

  tsp.grow(yfm2_tsp, e_tsp)
  tsp.regularize()
  emi_yfm2 = tsp.emi()

  tsp.grow(yfm3_tsp, e_tsp)
  tsp.regularize()
  emi_yfm3 = tsp.emi()

  # tsp.grow(yfm4_tsp, e_tsp)
  # tsp.regularize()
  # emi_yfm4 = tsp.emi()


  tsp.grow(ym1_tsp, e_tsp)
  tsp.regularize()
  emi_ym1 = tsp.emi()

  tsp.grow(ym2_tsp, e_tsp)
  tsp.regularize()
  emi_ym2 = tsp.emi()

  # tsp.grow(ym3_tsp, e_tsp)
  # tsp.regularize()
  # emi_ym3 = tsp.emi()

  # tsp.grow(ym4_tsp, e_tsp)
  # tsp.regularize()
  # emi_ym4 = tsp.emi()

  emis_fm1.append(emi_yfm1)
  emis_fm2.append(emi_yfm2)
  emis_fm3.append(emi_yfm3)
  # emis_fm4.append(emi_yfm4)


  emis_ym1.append(emi_ym1)
  emis_ym2.append(emi_ym2)
  # emis_ym3.append(emi_ym3)
  # emis_ym4.append(emi_ym4)

```

```python
window_size = int(3600/10*2) # 30 minutes



start = 5
stop = -1

y = df['hc'][start:stop]
y_hat = df['y_hat_1'][start:stop]

y_fm1 = df['f_pid'][start:stop]
# y_fm1 = (y_fm1-y_fm1.min())/(y_fm1.max()-y_fm1.min())

y_fm2 = df['freq_prev'][start:stop]
# y_fm2 = (y_fm2-y_fm2.min())/(y_fm2.max()-y_fm2.min())

y_fm3 = y_fm1-y_fm2  #df['freq_prev_3'][start:stop]
# y_fm3 = (y_fm3-y_fm3.min())/(y_fm3.max()-y_fm3.min())

# y_fm4 = df['freq_prev_4'][start:stop]
# y_fm4 = (y_fm4-y_fm4.min())/(y_fm4.max()-y_fm4.min())


y_m1 = df['hc_prev'][start:stop]
# y_m1 = (y_m1-y_m1.min())/(y_m1.max()-y_m1.min())

y_m2 = df['hc_prev_prev'][start:stop]
# y_m2 = (y_m2-y_m2.min())/(y_m2.max()-y_m2.min())

# y_m3 = df['hc_prev_3'][start:stop]
# y_m3 = (y_m3-y_m3.min())/(y_m3.max()-y_m3.min())

# y_m4 = df['hc_prev_4'][start:stop]
# y_m4 = (y_m4-y_m4.min())/(y_m4.max()-y_m4.min())


e = y-y_hat


emis_fm1 = [0] * window_size
emis_fm2 = [0] * window_size
emis_fm3 = [0] * window_size
# emis_fm4 = [0] * window_size

emis_ym1 = [0] * window_size
emis_ym2 = [0] * window_size
# emis_ym3 = [0] * window_size
# emis_ym4 = [0] * window_size

p_emis_fm1 = [0] * window_size
p_emis_fm2 = [0] * window_size
p_emis_fm3 = [0] * window_size
# emis_fm4 = [0] * window_size

p_emis_ym1 = [0] * window_size
p_emis_ym2 = [0] * window_size

for i in range(len(y)-window_size+1):
  # print(i)

  e_tsp = np.copy(np.array(e[i:i+window_size]), order='F')

  yfm1_tsp =  np.copy(np.array(y_fm1[i:i+window_size]), order='F')
  yfm2_tsp =  np.copy(np.array(y_fm2[i:i+window_size]), order='F')
  yfm3_tsp =  np.copy(np.array(y_fm3[i:i+window_size]), order='F')
  # yfm4_tsp =  np.copy(np.array(y_fm4[i:i+window_size]), order='F')


  ym1_tsp = np.copy(np.array(y_m1[i:i+window_size]), order='F')
  ym2_tsp = np.copy(np.array(y_m2[i:i+window_size]), order='F')
  # ym3_tsp = np.copy(np.array(y_m3[i:i+window_size]), order='F')
  # ym4_tsp = np.copy(np.array(y_m4[i:i+window_size]), order='F')

  #################################################


  emi_yfm1 = stats.pearsonr(e_tsp, yfm1_tsp)


  emi_yfm2 = stats.pearsonr(e_tsp, yfm2_tsp)


  emi_yfm3 = stats.pearsonr(e_tsp, yfm3_tsp)




  emi_ym1 = stats.pearsonr(e_tsp, ym1_tsp)


  emi_ym2 = stats.pearsonr(e_tsp, ym2_tsp)



  emis_fm1.append(emi_yfm1[0])
  emis_fm2.append(emi_yfm2[0])
  emis_fm3.append(emi_yfm3[0])
  # emis_fm4.append(emi_yfm4)


  emis_ym1.append(emi_ym1[0])
  emis_ym2.append(emi_ym2[0])
  # emis_ym3.append(emi_ym3)
  # emis_ym4.append(emi_ym4)

  p_emis_fm1.append(emi_yfm1[1])
  p_emis_fm2.append(emi_yfm2[1])
  p_emis_fm3.append(emi_yfm3[1])
  # emis_fm4.append(emi_yfm4)


  p_emis_ym1.append(emi_ym1[1])
  p_emis_ym2.append(emi_ym2[1])
```
